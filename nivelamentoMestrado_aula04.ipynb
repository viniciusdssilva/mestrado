```
#Definindo a função de leitura
import csv

def read_csv(filepath):
    table = list()  #Tabela é uma lista de linhas
    with open(filepath) as csv_file:
        reader = csv.DictReader(csv_file)
        for row in reader:
            table.append(row)
            
    return table                       
```

```
#Lendo o arquivo de dados sobre o Titanic
titanic_tbl = read_csv("train.csv")
titanic_tbl  
  
```

```
#Cálculo da média das idades
soma_idades = 0
contados = 0

for passg in titanic_tbl:
  try:
    soma_idades += float(passg["Age"])
    contados += 1
  except(ValueError):
    pass #O pass é uma instrução para indicar que não será feito nada. Um 'continue' faria sentido, mas, semanticamente, não é a abordagem mais adequada.
  
media = soma_idades / contados

media
```

```
#Cálculo da média das idades dos passageiros que morreram e dos que sobreviveram
soma_idades = 0
soma_idades_vivos = 0
soma_idades_mortos = 0

contados = 0
contados_vivos = 0
contados_mortos = 0

for passg in titanic_tbl:
  try:
    soma_idades += float(passg["Age"])
    if(passg["Survived"] == "1"):
      soma_idades_vivos += float(passg["Age"])
      contados_vivos += 1
    else:
      soma_idades_mortos += float(passg["Age"])
      contados_mortos += 1
    contados += 1
  except(ValueError):
    pass 
  
media = soma_idades / contados

media_vivos = soma_idades_vivos / contados_vivos

media_mortos = soma_idades_mortos / contados_mortos

print("Média de todas as idades:", "%.2f" %media) #String de formatação

print("Média dos sobreviventes:", "%.2f" %media_vivos)

print("Média dos mortos:", "%.2f" %media_mortos)
```

```
#Cálculo do percentual de sobreviventes
contados = 0
contados_vivos = 0

for passg in titanic_tbl:
  if(passg["Survived"] == "1"):
    contados_vivos += 1
  contados += 1
  
sobreviventes = contados_vivos / contados

print("Percentual de sobreviventes:", ("%.2f" %(sobreviventes*100)) + "%") #Gambiarra pra evitar o espaço inserido pela print entre as strings
```

```
#Soluções mais fáceis para os problemas do Titanic, utilizando compreensão de listas

def is_float(x):
  try:
    float(x)
    return True
  except:
    return False

idades = [float(passg["Age"]) for passg in titanic_tbl if is_float(passg["Age"])]

idades_vivos = [float(passg["Age"]) for passg in titanic_tbl if is_float(passg["Age"]) and passg["Survived"] == "1"]

idades_mortos = [float(passg["Age"]) for passg in titanic_tbl if is_float(passg["Age"]) and passg["Survived"] == "0"]


media = sum(idades) / len(idades)

media_vivos = sum(idades_vivos) / len(idades_vivos)

media_mortos = sum(idades_mortos) / len(idades_mortos)


print("Média de todas as idades:", "%.2f" %media)

print("Média dos sobreviventes:", "%.2f" %media_vivos)

print("Média dos mortos:", "%.2f" %media_mortos)

```
